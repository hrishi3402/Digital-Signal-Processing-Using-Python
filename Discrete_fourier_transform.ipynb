```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N =100

x = np.linspace(-2,2,N)
y = np.zeros(N)
```

```python
y[abs(x)<1] = x[abs(x)<1]
```

```python
plt.plot(x,y)
plt.show()
```

```python
c = np.zeros(N,complex)
n = np.arange(N)
```

```python
for k in range(N):
    c[k] = np.sum(y*np.exp(-2j*(np.pi)*n*k/N))
```

```python
plt.figure(figsize = (12,4))
plt.subplot(1,2,1)
plt.plot(abs(c), "r", label = "magnitude")
plt.legend()

plt.subplot(1,2,2)
plt.plot(np.angle(c), "r", label = "angle")
plt.legend()

plt.show()
```
