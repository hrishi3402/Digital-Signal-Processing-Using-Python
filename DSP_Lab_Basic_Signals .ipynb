Delta Function

```python
import numpy as np
import matplotlib.pyplot as plt


def delta(n):
    if n == 0:
        return 1
    else:
        return 0
```

```python
x = np.linspace(-7,7,num = 15)
y = [delta(n) for n in x]
plt.stem(x,y,use_line_collection = True)
plt.grid(True)
plt.show()
```

Unit Function

```python
x = np.linspace(-10,10,num = 1000)
y = [1 if n >=0 else 0 for n in x]
plt.plot(x,y,color='g')
plt.axhline(linewidth=1, color='b')
plt.axvline(linewidth=1, color='b')
plt.grid(True)
plt.show()
```

Rectangular Function

```python
x = np.linspace(-10,10,num=1000)
y = [4 if (n>=-2 and n<=2) else 0 for n in x]
plt.plot(x,y,color='r',linewidth='2')
plt.axhline(linewidth=1, color='b')
plt.axvline(linewidth=1, color='b')
plt.grid(True)
plt.show()
```

Sine Function

```python
x = np.linspace(-10,10,num=1000)
y = [np.sin(n) for n in x]
plt.stem(x,y,use_line_collection=True)

plt.grid(True)
plt.show()
```

Cosine Function

```python
x = np.linspace(-10,10,num=1000)
y = [np.cos(n) for n in x]
plt.stem(x,y,use_line_collection=True)
plt.axhline(linewidth=1, color='b')
plt.axvline(linewidth=1, color='b')
plt.grid(True)
plt.show()
```

Positive Exponential Function

```python
x = np.linspace(0,10,num=1000)
y = np.exp(x)
plt.plot(x,y,color='b',linewidth=2)
plt.axhline(linewidth=1, color='b')
plt.axvline(linewidth=1, color='b')
plt.grid(True)
plt.show()
```

Negative Exponential Function

```python
x = np.linspace(0,10,num=1000)
y = np.exp(-x)
plt.plot(x,y,color='b',linewidth=2)
plt.axhline(linewidth=1, color='b')
plt.axvline(linewidth=1, color='b')
plt.grid(True)
plt.show()
```
