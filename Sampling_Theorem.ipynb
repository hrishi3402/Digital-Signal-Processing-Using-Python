Over Sampling

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
f = 3
fs = 60
t = np.arange(-1,1,1/fs)
x = 2*np.sin(2*np.pi*f*t) + 3*np.cos(2*np.pi*f*t) 

ax.plot(t,x, 'd-')
ax.set_xlabel('time',fontsize=15)
ax.set_ylabel('amplitude',fontsize=15)
plt.title('Over Sampling')
```
